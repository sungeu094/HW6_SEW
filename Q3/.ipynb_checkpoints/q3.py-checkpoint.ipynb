```python
import csv
import matplotlib.pyplot as plt

def main():
    f = open('q3.csv', 'r', encoding='ANSI')
    data = csv.reader(f)
    header = next(data)

    total_male = 0
    total_female = 0

    for row in data:
        if '(5000000000)' in row[0]:
                total_male += int(row[1].replace(',', ''))
                total_female += int(row[104].replace(',', ''))
    
    
    f.close()
    
    labels = ['Male', 'Female']
    sizes = [total_male, total_female]
    colors = ['lightblue','pink']

    plt.title("Total Male and Female Populations in 2022")

    plt.barh('Female', total_female, color = colors[1], label="Female")
    plt.barh('Male', total_male, color = colors[0],label="Male")
    plt.xlabel('Population')
    plt.legend()
    plt.text(total_female, 'Female', f'{total_female:,}', va='center')
    plt.text(total_male, 'Male', f'{total_male:,}', va='center')
    plt.show()

    plt.pie(sizes, labels=labels,colors = colors, autopct='%1.3f%%')
    plt.show()


    f1 = open('q3(1).csv', 'r', encoding='ANSI')
    data1 = csv.reader(f1)
    header1 = next(data1)

    total_male1 = 0
    total_female1 = 0

    for row in data1:
        if '(5000000000)' in row[0]:
                total_male1 += int(row[1].replace(',', ''))
                total_female1 += int(row[104].replace(',', ''))


    f1.close()

    labels1 = ['Male', 'Female']
    sizes1 = [total_male1, total_female1]
    colors1 = ['lightgray','lightyellow']

    plt.title("Total Male and Female Populations in 2021")

    plt.barh('Female', total_female1, color = colors1[1], label="Female")
    plt.barh('Male', total_male1, color = colors1[0],label="Male")
    plt.xlabel('Population')
    plt.legend()
    plt.text(total_female1, 'Female', f'{total_female1:,}', va='center')
    plt.text(total_male1, 'Male', f'{total_male1:,}', va='center')
    plt.show()

    plt.pie(sizes1, labels=labels1,colors = colors1, autopct='%1.3f%%')
    plt.show()

    f2 = open('q3(2).csv', 'r', encoding='ANSI')
    data2 = csv.reader(f2)
    header2 = next(data2)

    total_male2 = 0
    total_female2 = 0

    for row in data2:
        if '(5000000000)' in row[0]:
                total_male2 += int(row[1].replace(',', ''))
                total_female2 += int(row[104].replace(',', ''))


    f2.close()

    labels2 = ['Male', 'Female']
    sizes2 = [total_male2, total_female2]
    colors2 = ['lime','lightcyan']

    plt.title("Total Male and Female Populations in 2020")

    plt.barh('Female', total_female2, color = colors2[1], label="Female")
    plt.barh('Male', total_male2, color = colors2[0],label="Male")
    plt.xlabel('Population')
    plt.legend()
    plt.text(total_female2, 'Female', f'{total_female2:,}', va='center')
    plt.text(total_male2, 'Male', f'{total_male2:,}', va='center')
    plt.show()

    plt.pie(sizes2, labels=labels2,colors = colors2, autopct='%1.3f%%')
    plt.show()

if __name__ == "__main__":
    main()


```
