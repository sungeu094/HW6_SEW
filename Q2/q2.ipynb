```python
import matplotlib.pyplot as plt
import csv
import numpy as np

def main():
    with open('q2.csv', mode='r') as file:
        reader = csv.reader(file)
        next(reader)
        count = []
        data = []
        for row in reader:
            count.append(int(row[0]))
            data.append([int(cell) for cell in row[1:]])

    fig, axs = plt.subplots(nrows=len(count), ncols=1, figsize=(8, 3 * len(count)))

    x_labels = ['1', '2', '3', '4', '5', '6']
    x = np.arange(len(x_labels))

    for i, ax in enumerate(axs):
        ax.hist(x + 0.5, bins=np.arange(len(x_labels) + 1), weights=data[i], label=str(count[i]), color='lightgray', alpha=0.7, rwidth=1.0)
        for j, v in enumerate(data[i]):
            ax.text(j + 0.5, v + 0.5, str(v), color='black', ha='center', va='center')
        ax.set_title('Dice Roll (Count={})'.format(count[i]))
        ax.legend(loc = 'lower center')
        ax.set_xticks(x + 0.5)
        ax.set_xticklabels(x_labels)

    plt.subplots_adjust(wspace=0.3, hspace=0.3)

    plt.show()

if __name__ == "__main__":
    main()


```
