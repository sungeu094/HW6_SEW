```python
import matplotlib.pyplot as plt

plt.title('plotting')          #제목은 상관없음. 범례를 위해 legend()를 사용하는 것
plt.plot([1,2,3,4], color = 'blue',label = 'asc') 
# Color를 통해 색깔 지정 가능, label을 통해 뭔지 이름 설정 가능
plt.plot([5,4,3,2], color = "gray", label = 'dasc')
plt.plot([2,3,4,5], 'r.', label = 'circle')      # r. -> r은 red .은 점을 찍는다는 표현
plt.plot([5,4,3,2], 'g^', label = 'triangle up')   #g^ -> g는 green ^는 위로 뾰족한 삼각형을 그린다는 표현
plt.legend() #legend가 범례가 나오도록 해주는 것이다.
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv', 'r', encoding = 'ANSI')
data = csv.reader(f)
header = next(data)
#print(header)
result = []
for row in data:
    if row[-1] != '':
        if row[0].split('-')[1] == '08':                #월을 바꾸고 싶으면 08만 교체
            result.append(float(row[-1]))               #최고 기온이 아닌 평균이나 최저 기온 원하면 -1 교체
f.close()

plt.title('Highest Temp in Seoul on August 2022')
plt.plot(result, 'hotpink')
plt.xlabel('days')
plt.ylabel('temperatures')
plt.show()
```

```python
#히스토그램
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)
aug = []

for row in data:
    month = row[0].split('-')[1]
    if row[-1] != '':
        if month == '08':
            aug.append(float(row[-1]))
            
plt.hist(aug, bins = 100, color = 'r')
plt.show()
```

```python
import matplotlib.pyplot as plt
data = [10,30,20,45]
label = ['A', 'B', 'C', 'D']
color = ['hotpink', 'darkmagenta', 'pink', 'deeppink']

plt.pie(data, labels = label, autopct = '%.1f%%', colors = color, explode =(0,0,0.1,0)) 
#C explode, autopct -> pie안에닫가 몇 퍼센트인지 쓸 수 있음(여기서는 소수점 첫째자리까지)
plt.legend()
plt.show()

xdata = [1,2,3,4]
ydata = [10,40,20,30]
sizes = [10,5000,100,200]
colors = ['red', 'blue', 'green', 'purple']  # 원하는 색깔로 변경

plt.scatter(xdata, ydata, s=sizes, c=colors)
plt.colorbar()
plt.show()
plt.show()
```

```python
import matplotlib.pyplot as plt
import csv

# 1. 파일을 올바르게 열기 위해 with 문 사용
with open('seoul.csv') as f:
    data = csv.reader(f)
    next(data)

    month = [[],[],[],[],[],[],[],[],[],[],[],[]]

    for row in data: 
        if row[-1] != '':
            # 3. 날짜와 온도 값이 제대로 매핑되지 않는 문제 해결
            date = row[0].split('-')
            month[int(date[1])-1].append(float(row[-1]))

# 2. 파일을 읽은 후에는 파일을 닫음
f.close()

# 5. 상자 그림의 축 레이블 설정
plt.boxplot(month)
plt.xlabel('Month')  # x축 레이블
plt.ylabel('Temperature (°C)')  # y축 레이블
plt.title('Monthly Temperature Distribution')  # 그래프 제목
plt.show()
```
